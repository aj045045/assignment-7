### Question 9

```python
listVal = {
'2014': pd.Series([100.5,12000,2000,50000],index=['Kusum','Kinshuk','Ankit','Shruti']),
'2015': pd.Series([150.8,18000,50000,60000],index=['Kusum','Kinshuk','Ankit','Shruti']),
'2016': pd.Series([200.9,22000,70000,70000],index=['Kusum','Kinshuk','Ankit','Shruti']),
'2017': pd.Series([30000,30000,10000,80000],index=['Kusum','Kinshuk','Ankit','Shruti']),
'2018': pd.Series([40000,45000,125000,90000],index=['Kusum','Kinshuk','Ankit','Shruti'])    
}
listVal = pd.DataFrame(listVal)
print(listVal)
```

### Question 10

```python
print(f"\nQuestion A :\n{listVal.index}")
print(f"\nQuestion B :\n{listVal.columns}")
print(f"\nQuestion C :\n{listVal.dtypes}")
print(f"\nQuestion D :\n")
print("Dimensions:", listVal.ndim)
print("Shape:", listVal.shape)
print("Size:", listVal.size)
print("Values:\n", listVal.values)

print(f"\nQuestion E :\n{listVal.tail(2)}")
print(f"\nQuestion F :\n{listVal.iloc[:, :2]}")

data = {
    '2018': pd.Series([160000, 110000, 500000, 340000, 900000], index=['Madhu', 'Kusum', 'Kinshuk', 'Ankit', 'Shruti'])
}
Sales2 = pd.DataFrame(data)
print(f"\nQuestion G :\n{Sales2}")
print(f"\nQuestion H :\n{Sales2.empty}")

```

### Question 11

```python
# a) Append Sales2 to Sales
Sales = pd.concat([listVal, Sales2], axis=0)
print("After appending Sales2:\n", Sales)

# b) Transpose Sales
Sales = Sales.T
print("After transposing:\n", Sales)

# c) Sales in 2017
print("Sales in 2017:\n", Sales.loc['2017'])

# d) Sales by Madhu and Ankit in 2017 and 2018
print("Sales by Madhu and Ankit in 2017 and 2018:\n", Sales.loc[['2017', '2018'], ['Madhu', 'Ankit']])

# e) Sales by Shruti in 2016
print("Sales by Shruti in 2016:\n", Sales.loc['2016', 'Shruti'])

# f) Add data for Sumeet
Sumeet_data = [196.2, 37800, 52000, 78438, 38852]
Sales.loc['Sumeet'] = Sumeet_data
print("After adding Sumeet's data:\n", Sales)

# g) Delete the data for the year 2014
Sales = Sales.drop('2014', axis=0)
print("After deleting the year 2014:\n", Sales)

# h) Delete the data for Kinshuk
Sales = Sales.drop('Kinshuk', axis=1)
print("After deleting Kinshuk's data:\n", Sales)

# i) Rename Ankit to Vivaan and Madhu to Shailesh
Sales = Sales.rename(columns={'Ankit': 'Vivaan', 'Madhu': 'Shailesh'})
print("After renaming:\n", Sales)

# j) Update Shailesh's sale in 2018 to 100000
Sales.loc['2018', 'Shailesh'] = 100000
print("After updating Shailesh's sale in 2018:\n", Sales)

# k) Write Sales to a CSV file without row and column labels
Sales.to_csv('SalesFigures.csv', header=False, index=False)

# l) Read the CSV file into SalesRetrieved and update labels
SalesRetrieved = pd.read_csv('SalesFigures.csv', header=None)
SalesRetrieved.index = Sales.index
SalesRetrieved.columns = Sales.columns
print("SalesRetrieved DataFrame:\n", SalesRetrieved)
```
